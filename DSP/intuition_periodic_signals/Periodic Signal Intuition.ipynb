```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# This is the time range

```python
t = np.arange(start=0,stop=20,step=0.1)
```

# The first Signal x1:
> X1 is a sinusoidal signal with frequency 0.1, which means the time peroid will be T = 1 / 0.1
<br>
> This means that the signal will oscillate 2 times in the range specified

```python
x1 = np.sin(2*np.pi*0.1*t)
```

```python
plt.plot(t,x1)
plt.grid()
```

# The Second Signal x2:
> X1 is a sinusoidal signal with frequency 0.2, which means the time peroid will be T = 1 / 0.2
<br>
> This means that the signal will oscillate 4 times in the range specified

```python
x2 = np.sin(2*np.pi*0.2*t)
```

```python
plt.plot(t,x2)
plt.grid()
```

# Humor me with some intuitions here
The first signal x1 repeats itself after - 10 units
<br>
The second signal x2 repeats itself after - 5 units
> Now superimposing the two signals will revela a little secret

```python
plt.plot(t,x1,t,x2)
plt.gca().legend(('x1','x2'))
plt.grid()
```

> Both the signals are periodic, and they have a common multiple too
<br>
> This multiple is the point where they both repeat themselves
<br>
> When the amplitudes of the two signals will be superimposed on themselves, the `common multiple point` is where the new signal will repeat itself, the other points can have `constructive` interference of `destructive` interference and hence their individual periodic points are of least interest.

```python
x = x1+x2
plt.plot(t,x)
plt.grid()
```

# Conclusion:
This is the sole reason as to why we find out the least common multiple of the two periodic signals and try to obtain the period of the new signal.
